```python
class UNet(nn.Module):
    def __init__(self):
        super(UNet, self).__init__()

        self.encoder1 = nn.Sequential(
            nn.Conv2d(3, 64, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            nn.Conv2d(64, 64, kernel_size=3, padding=1),
            nn.ReLU(inplace=True)
        )
        self.pool1 = nn.MaxPool2d(kernel_size=2)

        self.encoder2 = nn.Sequential(
            nn.Conv2d(64, 128, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            nn.Conv2d(128, 128, kernel_size=3, padding=1),
            nn.ReLU(inplace=True)
        )
        self.pool2 = nn.MaxPool2d(kernel_size=2)

        self.encoder3 = nn.Sequential(
            nn.Conv2d(128, 256, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            nn.Conv2d(256, 256, kernel_size=3, padding=1),
            nn.ReLU(inplace=True)
        )
        self.pool3 = nn.MaxPool2d(kernel_size=2)

        self.encoder4 = nn.Sequential(
            nn.Conv2d(256, 512, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            nn.Conv2d(512, 512, kernel_size=3, padding=1),
            nn.ReLU(inplace=True)
        )
        self.pool4 = nn.MaxPool2d(kernel_size=2)

        self.bottleneck = nn.Sequential(
            nn.Conv2d(512, 1024, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            nn.Conv2d(1024, 1024, kernel_size=3, padding=1),
            nn.ReLU(inplace=True)
        )

        self.up4 = nn.ConvTranspose2d(1024, 512, kernel_size=2, stride=2)
        self.decoder4 = nn.Sequential(
            nn.Conv2d(1024, 512, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            nn.Conv2d(512, 512, kernel_size=3, padding=1),
            nn.ReLU(inplace=True)
        )

        self.up3 = nn.ConvTranspose2d(512, 256, kernel_size=2, stride=2)
        self.decoder3 = nn.Sequential(
            nn.Conv2d(512, 256, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            nn.Conv2d(256, 256, kernel_size=3, padding=1),
            nn.ReLU(inplace=True)
        )

        self.up2 = nn.ConvTranspose2d(256, 128, kernel_size=2, stride=2)
        self.decoder2 = nn.Sequential(
            nn.Conv2d(256, 128, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            nn.Conv2d(128, 128, kernel_size=3, padding=1),
            nn.ReLU(inplace=True)
        )

        self.up1 = nn.ConvTranspose2d(128, 64, kernel_size=2, stride=2)
        self.decoder1 = nn.Sequential(
            nn.Conv2d(128, 64, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            nn.Conv2d(64, 64, kernel_size=3, padding=1),
            nn.ReLU(inplace=True)
        )

        self.final = nn.Conv2d(64, 1, kernel_size=1)

    def forward(self, x):
        conv1 = self.encoder1(x)
        pool1 = self.pool1(conv1)

        conv2 = self.encoder2(pool1)
        pool2 = self.pool2(conv2)

        conv3 = self.encoder3(pool2)
        pool3 = self.pool3(conv3)

        conv4 = self.encoder4(pool3)
        pool4 = self.pool4(conv4)

        bottleneck = self.bottleneck(pool4)

        up4 = self.up4(bottleneck)
        up4 = torch.cat([up4, conv4], dim=1)
        dec4 = self.decoder4(up4)

        up3 = self.up3(dec4)
        up3 = torch.cat([up3, conv3], dim=1)
        dec3 = self.decoder3(up3)

        up2 = self.up2(dec3)
        up2 = torch.cat([up2, conv2], dim=1)
        dec2 = self.decoder2(up2)

        up1 = self.up1(dec2)
        up1 = torch.cat([up1, conv1], dim=1)
        dec1 = self.decoder1(up1)

        out = self.final(dec1)
        out = torch.sigmoid(out)
        return out

class LaneDataset(Dataset):
    def __init__(self, image_dir, mask_dir, dim=(128, 128)):
        self.image_dir = image_dir
        self.mask_dir = mask_dir
        self.dim = dim
        self.image_ids = [os.path.splitext(file)[0] for file in os.listdir(image_dir) if file.endswith('.jpg')]
        # self.image_ids = random.sample(self.image_ids, 1000)

    def __len__(self):
        return len(self.image_ids)

    def __getitem__(self, index):
        ID = self.image_ids[index]
        image_path = os.path.join(self.image_dir, ID + '.jpg')
        mask_path = os.path.join(self.mask_dir, ID + '.jpg.json')

        image = cv2.imread(image_path)
        original_shape = image.shape[:2]
        image = cv2.resize(image, self.dim)
        image = image.transpose((2, 0, 1)) / 255.0

        # print("here1")
        mask = np.zeros(original_shape)
        with open(mask_path, 'r') as file:
            data = json.load(file)
            for obj in data['objects']:
                if obj['classTitle'] == 'lane':
                    points = obj['points']['exterior']
                    cv2.line(mask, tuple(points[0]), tuple(points[1]), (255), thickness=5)
        # print(np.unique(mask))
        mask = cv2.resize(mask, self.dim)
        _, mask = cv2.threshold(mask, 0, 255, cv2.THRESH_BINARY)
        # print(np.unique(mask))
        mask = np.expand_dims(mask, axis=0) / 255.0
        # print(np.unique(mask))

        # print("here2")
        return torch.tensor(image, dtype=torch.float32), torch.tensor(mask, dtype=torch.float32), image_path, mask_path

def iou(outputs, labels):
    # Предполагается, что outputs и labels - тензоры размером (batch_size, channels, height, width)
    # И оба бинаризированы, где объекты обозначаются единицами.
    intersection = (outputs & labels).float().sum((1, 2, 3))  # Пересечение
    union = (outputs | labels).float().sum((1, 2, 3))         # Объединение

    iou = (intersection + 1e-6) / (union + 1e-6)  # добавляем небольшое число, чтобы избежать деления на ноль
    return iou.mean()
```

```python
# Пути к вашим данным
train_image_dir = './data/bdd100k:-images-100k/train/img'
train_mask_dir = './data/bdd100k:-images-100k/train/ann'
val_image_dir = './data/bdd100k:-images-100k/val/img'
val_mask_dir = './data/bdd100k:-images-100k/val/ann'

train_dataset = LaneDataset(train_image_dir, train_mask_dir)
val_dataset = LaneDataset(val_image_dir, val_mask_dir)

train_loader = DataLoader(train_dataset, batch_size=16, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=16, shuffle=False)
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = UNet().to(device)
criterion = nn.BCELoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

num_epochs = 10
patience = 10
best_loss = float('inf')
patience_counter = 0

for epoch in range(num_epochs):
    print("start")
    model.train()
    train_loss = 0
    train_iou = 0
    for images, masks, _, _ in tqdm(train_loader, desc=f'Epoch {epoch+1}/{num_epochs}'):
        images, masks = images.to(device), masks.to(device)
        outputs = model(images)
        loss = criterion(outputs, masks)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        predicted_masks = outputs > 0.5
        iou_score = iou(predicted_masks, masks > 0.5)

        train_loss += loss.item() * images.size(0)
        train_iou += iou_score.item() * images.size(0)

    train_loss /= len(train_loader.dataset)
    train_iou /= len(train_loader.dataset)

    model.eval()
    val_loss = 0
    val_iou = 0
    with torch.no_grad():
        for images, masks, _, _ in val_loader:
            images, masks = images.to(device), masks.to(device)

            outputs = model(images)
            loss = criterion(outputs, masks)

            predicted_masks = outputs > 0.5
            # print(np.unique(outputs))
            # print(predicted_masks)
            # print(masks)
            iou_score = iou(predicted_masks, masks > 0.5)

            val_loss += loss.item() * images.size(0)
            val_iou += iou_score.item() * images.size(0)

    val_loss /= len(val_loader.dataset)
    val_iou /= len(val_loader.dataset)

    print(f'Epoch {epoch+1}/{num_epochs}, Train Loss: {train_loss}, Train IoN: {train_iou}, Validation Loss: {val_loss}, Validation IoU: {val_iou}')

    if val_loss < best_loss:
        best_loss = val_loss
        patience_counter = 0
    else:
        patience_counter += 1

    if patience_counter >= patience:
        print("Early stopping")
        break
```
